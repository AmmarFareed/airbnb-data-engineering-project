```python
import pandas as pd
from scipy.stats import mannwhitneyu

df = pd.read_csv("../data/processed/listings_featured.csv")
```

## Statistical Analysis

This section applies statistical testing to validate key insights observed during exploratory data analysis.

The goal is to determine whether observed patterns are statistically significant.
``

```python
entire = df[df['room_type'] == 'Entire home/apt']['price']
private = df[df['room_type'] == 'Private room']['price']

entire = entire[entire < 1000]
private = private[private < 1000]

stat, p_value = mannwhitneyu(entire, private)

print("Statistic:", stat)
print("P-value:", p_value)
```

### Hypothesis Test: Entire Home vs Private Room Pricing

**Objective:**  
To determine whether entire homes are priced significantly higher than private rooms.

**Null Hypothesis (H0):**  
There is no difference in price between entire homes and private rooms.

**Alternative Hypothesis (H1):**  
There is a significant difference in price between the two room types.

**Method:**  
The Mann-Whitney U test was used due to the non-normal distribution of price data.

**Result:**  
The test returned a p-value approximately equal to 0, indicating a statistically significant difference between the two groups.

**Conclusion:**  
We reject the null hypothesis and conclude that entire homes are significantly more expensive than private rooms.

### Business Meaning:

This confirms that guests are willing to pay a premium for privacy and exclusive access.

Hosts offering entire homes can adopt higher pricing strategies, while private room hosts must compete more on affordability.

This insight validates pricing strategies observed in the exploratory data analysis.

Given the extremely small p-value, the difference is not only statistically significant but also highly robust.

```python
superhost = df[df['host_is_superhost'] == 't']['review_scores_rating']
non_superhost = df[df['host_is_superhost'] == 'f']['review_scores_rating']

# remove zeros (no reviews)
superhost = superhost[superhost > 0]
non_superhost = non_superhost[non_superhost > 0]

stat, p_value = mannwhitneyu(superhost, non_superhost)

print("Statistic:", stat)
print("P-value:", p_value)
```

### Hypothesis Test: Superhost vs Review Rating

**Objective:**  
To determine whether superhosts receive higher review ratings compared to non-superhosts.

**Null Hypothesis (H0):**  
There is no difference in review ratings between superhosts and non-superhosts.

**Alternative Hypothesis (H1):**  
There is a significant difference in review ratings between the two groups.

**Method:**  
The Mann-Whitney U test was used due to the non-normal distribution of rating data.  
Listings without reviews were excluded to ensure valid comparisons.

**Result:**  
The test returned a p-value approximately equal to 0, indicating a statistically significant difference between the two groups.

**Conclusion:**  
We reject the null hypothesis and conclude that superhosts have significantly different (higher) review ratings compared to non-superhosts.

### Business Meaning:

This confirms that superhost status is associated with higher service quality and better guest experiences.

Superhosts likely maintain higher standards in communication, cleanliness, and reliability, which translates into better reviews.

For new hosts, improving service consistency and customer experience can increase the chances of achieving superhost status and attracting more bookings.


This suggests that rating systems and platform incentives effectively differentiate higher-quality hosts in the marketplace.

```python
from scipy.stats import f_oneway

# select top neighbourhoods
top_neigh = df[df['price'] < 1000]['neighbourhood_cleansed'].value_counts().head(5).index

groups = [
    df[(df['neighbourhood_cleansed'] == n) & (df['price'] < 1000)]['price']
    for n in top_neigh
]

stat, p_value = f_oneway(*groups)

print("Statistic:", stat)
print("P-value:", p_value)
```

### Hypothesis Test: Price Differences Across Neighbourhoods

**Objective:**  
To determine whether average listing prices differ significantly across neighbourhoods.

**Null Hypothesis (H0):**  
All neighbourhoods have the same average price.

**Alternative Hypothesis (H1):**  
At least one neighbourhood has a different average price.

**Method:**  
A one-way ANOVA test was used to compare mean prices across multiple neighbourhood groups.

**Result:**  
The test returned a p-value approximately equal to 0, indicating a statistically significant difference among neighbourhoods.

**Conclusion:**  
We reject the null hypothesis and conclude that listing prices vary significantly across neighbourhoods.

### Business Meaning:

This confirms that location is one of the strongest drivers of pricing in the Airbnb market.

Neighbourhood characteristics such as proximity to attractions, business districts, and amenities significantly influence pricing power.

For hosts, pricing strategies must be tailored to the specific neighbourhood.  

For investors, selecting properties in high-price neighbourhoods can lead to greater revenue potential.


This result validates the spatial pricing patterns observed during exploratory data analysis and highlights the importance of location-specific market dynamics.
